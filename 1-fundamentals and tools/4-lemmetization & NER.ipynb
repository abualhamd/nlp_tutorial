```python
import nltk
from nltk.stem import WordNetLemmatizer
import spacy
```

```python
from nltk.corpus import state_union

text = state_union.raw('2006-GWBush.txt')
print(text)
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
doc = nlp(text)

for token in doc[70:]:
    print('{0:20}{1:20}'.format(token.text, token.lemma_))
```

```python
wordnet_stemmer = WordNetLemmatizer()
tokenized_text = nltk.word_tokenize(text)

for token in tokenized_text[68:]:
    print('{0:20}{1:20}'.format(token, wordnet_stemmer.lemmatize(token)))
```

```python
for ent in doc.ents:
    print('{0:40}{1:20}{2:20}'.format(ent.text, ent.label_, spacy.explain(ent.label_)))
```

```python
for chunk in doc.noun_chunks:
    print('{0:40}{1:20}{2:20}{3:20}'.format(chunk.text, chunk.root.text, chunk.root.dep_, spacy.explain(chunk.root.dep_)))
```
